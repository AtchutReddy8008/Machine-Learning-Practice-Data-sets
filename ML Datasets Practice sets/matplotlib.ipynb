```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 15]

# Create a plot
plt.plot(x, y)

# Add a title and labels
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()

```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 15]

# Create a scatter plot
plt.scatter(x, y)

# Add a title and labels
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 4]

# Create a bar plot
plt.bar(categories, values)

# Add a title and labels
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]

# Create a histogram
plt.hist(data, bins=5, edgecolor='black')

# Add a title and labels
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title("Pie Chart")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [15, 25, 35, 45, 55]

plt.plot(x, y1, linestyle='--', color='r', marker='o', label='Line 1')
plt.plot(x, y2, linestyle='-', color='b', marker='x', label='Line 2')

plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)

# Save the plot
plt.savefig('custom_plot.png')

# Show the plot
plt.show()
```
