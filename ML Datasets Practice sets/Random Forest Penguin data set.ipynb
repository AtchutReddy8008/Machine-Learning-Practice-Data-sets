```python
import pandas as pd
import seaborn as sns

# Load the penguins dataset
df = sns.load_dataset("penguins")

# Create a copy of the DataFrame
df2 = df.copy()

# Check the value counts of the "sex" column
sex_counts = df2["sex"].value_counts()

print(sex_counts)
```

### Feature Engineering

```python
df.sex.unique()
```

```python
df.sex.value_counts()
```

```python
k=pd.get_dummies(df.sex,drop_first=True)
```

```python
df.island.unique()
```

```python
m=pd.get_dummies(df.island,drop_first=True)
```

```python
newdf=pd.concat([df,m,k])
```

```python
newdf
```

```python
newdf.drop(["sex","island"],axis=1,inplace=True)
```

```python
newdf
```

```python
Y=newdf.species
Y.head()
```

```python
Y.unique()
```

```python
Y=Y.map({"Adelie":0,'Chinstrap':1,'Gentoo':2})
```

```python
Y.head()
```

```python
X=newdf
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Training random Forest Classification on Training Set

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=5,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```
